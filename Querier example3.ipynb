### Merge revision related tables with tokens of all_content

```python
import pandas as pd
from wikiwho_wrapper import WikiWhoAPI, APIQuerier

api = WikiWhoAPI()

querier = APIQuerier(api)

all_tokens_history = querier.all_content("bioglass")
last_revision = querier.last_rev_content("bioglass")
```

```python
tokens_history_of_last_revision = pd.merge(last_revision, all_tokens_history[['token_id','in', 'out']], how='left', on='token_id')
```
